```python
df_numeral= df2[["PassengerId","Pclass","SibSp","Parch","New_embark","Sex_type","cabin_allot","Fareno.","age_type"]]
```

```python
df_x =df_numeral
df_y = df2["Survived"]
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test =train_test_split(df_x,df_y,test_size =0.2)
score_list=[]
```

```python
#Linear REGRESSION Model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model = lr.fit(X_train,Y_train )
score_lr = model.score(X_test,Y_test)
y_lr_predic = model.predict(X_test)
score_list.append(score_lr)
```

```python
#Logistic Regression model 
from sklearn.linear_model import LogisticRegression
lor= LogisticRegression(solver='liblinear', random_state=10)
lomodel = lor.fit(X_train,Y_train)
score_lor = lomodel.score(X_test,Y_test)
y_lor_predic = lomodel.predict(X_test)
score_list.append(score_lor)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=6)
modelDT = dt.fit(X_train,Y_train)
dt_score = modelDT.score(X_test,Y_test)
y_dt_predic = modelDT.predict(X_test)
score_list.append(dt_score)
```

```python
from sklearn.svm import SVC # "Support Vector Classifier" 
svc = SVC(kernel='linear')
modelSV = svc.fit(X_train,Y_train)
y_sv_predic = modelSV.predict(X_test)
sv_score = modelSV.score(X_test,Y_test)
score_list.append(sv_score)
```

```python
model_name = ["LinearRegression" , "LogisticRegression" , "Decision Tree" , "SVMClassifier"]
dic_model= {"name":model_name , "Score":score_list }
print(dic_model)
```
