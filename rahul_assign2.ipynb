```python
empty_list = ["Nohashtags"]
dataset["hashtags"].fillna("None", inplace = True)
```

```python
h_series = dataset["hashtags"]
hashtags_list = h_series.tolist()
```

```python
h= []
for i in hashtags_list : 
    if i == "None" :
        h.append(empty_list)
    else :
        h.append(i)
        
```

```python
for i in range(20) :
    print(type(h[i]))
```

```python
new_h=[]

for i in h :
    new=[]
    if isinstance(i, list):
        for j in i :
            x =j.lower()
            new.append(x)
    else :
        x =i.lower()
        new.append(x)
    new_h.append(new)
        
    
    
        
```

```python
for i in range(20) :
    print(new_h[i])
    print(type(new_h[i]))
```

```python
import re 
cov_match=[]
for i in new_h : 
    for j in i : 
        if re.search("covi|coro|covid|virus",j):
            cov_match.append(1)
        else :
            cov_match.append(0)
        
```

```python
dataset.drop(["new_column"],axis=1)
```

```python
text_s = dataset["text"]
text_li = text_s.tolist()
new=[]
for i in text_li : 
    w = i.lower()
    new.append(w)
dataset["new"] = new
text_match = []
for i in new : 
    if re.search(r"covid|corona",i):
        text_match.append(1)
    else :
        text_match.append(0)
dataset =dataset.drop(["new"],axis=1)       
```

```python
for i in range(0,10) :
    print(text_match[i])

dataset =dataset.drop(["new_column"],axis = 1)
```

```python
dataset.columns
```

```python
dataset["covid_hashtags"] = cov_match
dataset["covid_text"] = text_match 
dataset.head(20)
```

```python
#filtering data only with covid19 tweets or texts
covid_data_1 = dataset.loc[dataset["covid_hashtags"]==1 ]
covid_data_2 = dataset.loc[dataset["covid_text"]==1]
```

```python
covid_data =  pd.concat([covid_data_1,covid_data_2],ignore_index=True)
covid_data.drop_duplicates(subset="user_name",keep="first",inplace=True)
```

```python
covid_data.head()
covid_data.shape
```

```python
loca = covid_data["user_location"] 
location=loca.tolist()
locat = set(loca)
dict_loc = {}

for i in locat :
    x= location.count(i)
    dict_loc[i] = x 
#print(dict_loc)
    
```

```python
import numpy as np
covid_data[['user_followers', 'user_friends', 'user_favourites']].max()
#mapping data to get a better view 
covid_data["followers"] = np.select([(covid_data["user_followers"]<=2500),(covid_data["user_followers"]>2500)&(covid_data["user_followers"]<=10000) ,(covid_data["user_followers"]>10000)&(covid_data["user_followers"]<=100000), (covid_data["user_followers"]>100000)&(covid_data["user_followers"]<=1000000),(covid_data["user_followers"]>1000000)]
,["less than 2500","<10000","<100000","less than a million","more than a million"])

covid_data["friends"] = np.select([(covid_data["user_friends"]<=2500),(covid_data["user_friends"]>2500)&(covid_data["user_friends"]<=10000) ,(covid_data["user_friends"]>10000)&(covid_data["user_friends"]<=100000), (covid_data["user_friends"]>100000)&(covid_data["user_friends"]<=1000000)]
,[1,2,3,4])

covid_data["favourites"] = np.select([(covid_data["user_favourites"]<=2500),(covid_data["user_favourites"]>2500)&(covid_data["user_favourites"]<=10000) ,(covid_data["user_favourites"]>10000)&(covid_data["user_favourites"]<=100000), (covid_data["user_favourites"]>100000)&(covid_data["user_favourites"]<=1000000),(covid_data["user_favourites"]>1000000)],[1,2,3,4,5])
```
